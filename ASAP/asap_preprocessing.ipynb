```python
import pandas as pd
from pathlib import Path
```

```python
input_path = Path(r"..\dataset\asap\training_set_rel3.xls")
output_path = Path(r"..\dataset\asap\training_set_rel3_preprocessed.csv")
```

```python
df = pd.read_excel(input_path)

print("Original shape:", df.shape)
print("Original columns:")
print(df.columns.tolist())
```

```python
required_cols = [
    "essay_id",
    "essay_set",
    "essay",
    "domain1_score"
]

df_processed = df[required_cols].copy()
```

```python
df_processed["essay"] = (
    df_processed["essay"]
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
)
```

```python
df_processed["essay_id"] = df_processed["essay_id"].astype("Int64")
df_processed["essay_set"] = df_processed["essay_set"].astype("Int64")
df_processed["domain1_score"] = df_processed["domain1_score"].astype("Int64")
```

```python
print("\nFinal shape:", df_processed.shape)
print("Final columns:")
print(df_processed.columns.tolist())

print("\nMissing values:")
print(df_processed.isna().sum())

print("\nSample:")
display(df_processed.head())
```

```python
df_processed.to_csv(output_path, index=False, encoding="utf-8")

print(f"\nSaved to: {output_path}")
```
